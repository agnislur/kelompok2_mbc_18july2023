Buatlah kode program dalam bahasa Python yang menerima satu inputan angka. Kemudian program akan menampilkan pesan apakah angka tersebut merupakan bilangan prima atau bukan bilangan prima.

============================================================================

Hasil program :

image.png

```python
print("## cek angka prima ##")
print("===============")
print("")

x = int(input("Input 1 angka bulat: "))
prima= True

if (x == 0) or (x == 1):
  prima = False
else:
  for a in range(2, x):
    if (x % a == 0):
      prima = False
      break

if(prima):
  print('{} adalah angka prima'.format(x))
else:
  print("{} bukan angka prima".format(x))
```

Buatlah program yang meminta data inputan berupa total belanja. Program kemudian menghitung berapa diskon yang bisa diperoleh dengan ketentuan sebagai berikut:

*   Jika total belanja kurang dari Rp. 100.000, tidak mendapat diskon.
*   Jika total belanja antara Rp. 100.000  Rp. 500.000, mendapat diskon 10%.
*   Jika total belanja antara Rp. 500.000  Rp. 1.000.000, mendapat diskon 20%.
*   Jika total belanja diatas Rp.1.000.000, mendapat diskon 30%.

Kode program kemudian menampilkan harga yang harus dibayar setelah dikurangi diskon.

===============================================================================

Hasil program

image.png

```python
print("## Program Python Diskon Potongan Harga ##")
print("==========================================")
print("")

total = int(input("Total Belanja: Rp."))

if (total <= 100000):
  hasil = total
  print("Anda tidak mendapatkan diskon")
elif (total >= 100000 and total <= 500000):
  total1 = (total * 0.1)
  hasil = (total - total1)
  print("Selamat, anda mendapat diskon 10%")
elif (total >= 500000 and total <= 1000000):
  total1 = (total * 0.2)
  hasil = (total - total1)
  print("Selamat, anda mendapat diskon 20%")
elif (total >= 1000000):
  total1 = (total * 0.3)
  hasil = (total - total1)
  print("Selamat, anda mendapat diskon 30%")
else:
  hasil = total


print("Total bayar: Rp.", hasil)
```

Buatlah program dalam bahasa Python untuk menentukan gaji karyawan mingguan dengan ketentuan sebagai berikut:

*   Golongan = A maka upah per jam 5000
*   Golongan = B maka upah per jam 7000
*   Golongan = C maka upah per jam 8000
*   Golongan = D maka upah per jam 10000

Ketentuan tambahan :


*   Jika jam kerja karyawan lebih dari 48 jam per minggu maka akan mendapat uang lembur dengan perhitungan uang lembur = (jam kerja-48)*4000.
*   Jika jam kerja kurang dari 48 jam maka pegawai tidak akan mendapat uang lembur.
Perhitungan gaji pegawai adalah upah + uang lembur.
*   Input berupa nama karyawan, golongan dan jam kerja.
*   Outputnya adalah nama karyawan dan gaji yang diterima.

Hasil program

image.png

```python
print("## Program Python Menghitung Gaji Karyawan ##")
print("=============================================")
print("")
# Daftar upah per jam sesuai golongan
upah_per_jam = {
    'A': 5000,
    'B': 7000,
    'C': 8000,
    'D': 10000
}

# Input data karyawan
nama_karyawan = input("Nama Karyawan: ")
golongan = input("Golongan: ")
jam_kerja = int(input("Jumlah Jam Kerja: "))

# Menghitung gaji
upah = upah_per_jam[golongan] * jam_kerja
if jam_kerja > 48:
    uang_lembur = (jam_kerja - 48) * 4000
    gaji = upah + uang_lembur
else:
    uang_lembur = 0
    gaji = upah

# Output hasil
print("\n## Program Python Menghitung Gaji Karyawan ##")
print("======================================")
print("\nNama Karyawan:", nama_karyawan)
print("Golongan:", golongan)
print("Jumlah Jam Kerja:", jam_kerja)
print("")
print(nama_karyawan, "menerima upah Rp.", gaji, "per minggu")
```
